#Regression 題目
糖尿病 從sklearn dataset 取資料

經驗法則:欄位越少,問題就有可能越簡單

```python
import pandas as pd
from sklearn.datasets import load_diabetes
diabetes = load_diabetes(scaled=False)
datas = pd.DataFrame(diabetes["data"], columns = diabetes["feature_names"])
datas["ans"] = diabetes["target"]
datas
```

```python
from sklearn.model_selection import train_test_split

x = datas.drop("ans", axis=1)
y = datas["ans"]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1)
```

```python
from sklearn.tree import DecisionTreeClassifier
reg = DecisionTreeClassifier(max_depth=2)
reg.fit(x_train, y_train)
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
plt.figure(figsize=(10, 10))
plot_tree(reg, feature_names=x.columns, filled=True)
```

```python
#將結果帶入 進行預測值推斷(模型結果估測率)

pre = reg.predict(x_test)
```
